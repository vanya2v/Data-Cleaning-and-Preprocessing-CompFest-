````python
# Create Synthetic data with prompt

prompt = """Generate python code that produces a
realistic data set with 35 rows,
rows are different 35 product names of popular groceries products/brands sold in shops in Bahasa Indonesia,
variables containing normal distributions, price has negative linear correlations to units sold and discount has postive linear correlation to units sold,
product catalog data set including the product name,
include the datetime as a column between 2022 and 2023,
make sure all arrays must all be same length

1. Popularity
2. Profit Margin
3. Seasonal Demand
4. Units sold
5. Availability
6. Price
7. Niche Market
8. Marketing

"""
MODEL = "gpt-3.5-turbo"
response = openai.ChatCompletion.create(
    model=MODEL,
    messages=[
        {"role": "user", "content": prompt},
    ],
    temperature=0,
)
print(response['choices'][0]['message']['content'])
````

Then, you can copy-paste the codes generated by GPT to the cell below to generate the pandas dataframe.

````python
import numpy as np
import pandas as pd
import random
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(0)

# Generate product names
product_name = ['Indomie', 'Bango', 'ABC', 'Sari Roti', 'Teh Pucuk', 'Pocari Sweat', 'Aqua', 'Mie Sedap', 'Teh Botol', 'Kecap Cap Bango',
                 'Kecap ABC', 'Susu Ultra', 'Susu Dancow', 'Susu Bendera', 'Susu Indomilk', 'Susu Milo', 'Susu Frisian Flag', 'Susu Bear Brand',
                 'Susu SGM', 'Susu Mom', 'Susu Nutrilon', 'Susu Enfamil', 'Susu Lactogen', 'Susu Bebelac', 'Susu Nestle', 'Susu Anmum',
                 'Susu Anlene', 'Susu Ensure', 'Susu Pediasure', 'Susu Greenfields', 'Susu Aptamil', 'Susu Neocate', 'Susu Isomil', 'Susu Alula',
                 'Susu Bebelove']

# Generate datetime column
# Generate datetime values between 2022 and 2023
start_date = datetime(2022, 1, 1)
end_date = datetime(2023, 7, 31)
date_range = end_date - start_date
date = [start_date + timedelta(days=random.randint(0, date_range.days)) for _ in range(35)]

# Generate variables with normal distributions
popularity = np.random.normal(loc=50, scale=10, size=35)
profit_margin = np.random.normal(loc=0.2, scale=0.05, size=35)
seasonal_demand = np.random.normal(loc=100, scale=20, size=35)
units_sold = np.random.normal(loc=500, scale=100, size=35)
availability = np.random.normal(loc=0.7, scale=0.1, size=35)
price = np.random.normal(loc=25000, scale=5000, size=35)
niche_market = np.random.normal(loc=0.5, scale=0.1, size=35)
marketing = np.random.normal(loc=50, scale=10, size=35)

# Generate discount with positive linear correlation to units sold
#discount = units_sold * (-0.02) + np.random.normal(loc=0, scale=5, size=35)

# Create product catalog dataframe
product_cat = pd.DataFrame({
    'Product Name': product_name,
    'Datetime': date ,
    'Popularity': popularity,
    'Profit Margin': profit_margin,
    'Seasonal Demand': seasonal_demand,
    'Units Sold': units_sold,
    'Availability': availability,
    'Price': price,
    'Niche Market': niche_market,
    'Marketing': marketing
    })


product_cat = product_cat.astype({'Units Sold':'int', 'Price':'int', 'Marketing':'int'})
product_cat = product_cat.round(decimals= 3)

# Print the product catalog dataframe and save to .csv file
print(product_cat)
product_cat.to_csv('product_cat.csv', index = False)
````
