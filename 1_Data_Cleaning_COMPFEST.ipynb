
# DATA CLEANING - STRUCTURED DATA

In this tutorial notebook, you will learn the basics of data cleaning that we had learned in the lecture session with a hands-on approach.


##  Study case:
A shop “Ratu Susu” has 35 items sold in the shop. The product catalogue dataset contains  features for predicting the best product to be sold in  “Ratu Susu” to get the optimum revenue. The dataset containing different 35 product names of popular groceries products/brands sold in "Ratu Susu" (mostly milk). Columns are features, such as price, units sold, etc., between 2022 and July 2023. As a data scientist, you have to clean the dataset first before modelling the solution.




```python
#read the raw product catalogue of "Ratu Susu"
import pandas as pd

product_catalogue = pd.read_csv("ratususu_products_raw.csv")
```

```python
#show the last 20 rows
product_catalogue.tail(20)
```

## **1. Remove duplicates / unwanted observations**


Check for duplicated entries and remove them.





```python
product_catalogue.duplicated() #check if there is any duplicate in rows

# Drop duplicated rows
product_catalogue = product_catalogue.drop_duplicates(ignore_index=True)
```

Remove irrelevant column in the data (feature 'Niche Market')



```python
# Drop the irrelevant column
product_catalogue = product_catalogue.drop(columns='Niche Market')
```

## **2. Change Data Types**

Useful to fix the structural error, types, and shape of the data before processing.

```python
# Check information, shape, and data type for the data frame
product_catalogue.info()
```

```python
# Convert column data type to integer
product_catalogue['Seasonal Demand'] = product_catalogue['Seasonal Demand'].astype('int')
```

## **3. Handling Missing Data**

Detect missing data (N/A, null, None) and handle how to replace/fill/interpolate them.

```python
#Return dataframe of Boolean values which are True for NaN values.
product_catalogue.isnull()
```

```python
# handling the missing values

# Option 1: Fill with 0 (or any specified value)
# product_catalogue = product_catalogue.fillna(0)

# Option 2: Interpolate the missing values
product_catalogue = product_catalogue.interpolate(method ='linear', limit_direction ='forward')
```

## **4. Data Transformation**

Converting the data from one form to another to make it more suitable for analysis.

```python
# scaling the profit margin to (0,1) with min-max scaler

from sklearn.preprocessing import MinMaxScaler

# initialising the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# scaling the profit margin to be 0 to 1
product_catalogue['Profit Margin'] = scaler.fit_transform(product_catalogue['Profit Margin'].values.reshape(-1, 1))
```

## **5. Managing Unwanted Outliers**

Detect the outliers in 'Price' column with IQR (Inter Quartile Range) and remove them


```python
# IQR
import numpy as np
Q1 = np.percentile(product_catalogue['Price'], 25, method='midpoint')
Q3 = np.percentile(product_catalogue['Price'], 75, method='midpoint')
IQR = Q3 - Q1
print('IQR:', IQR)

# Above Upper bound
upper=Q3+1.5*IQR
upper_array=np.array(product_catalogue['Price']>=upper)
print("Upper Bound:",upper)
print("#Outlier above upper bound: ", upper_array.sum())

#Below Lower bound
lower=Q1-1.5*IQR
lower_array=np.array(product_catalogue['Price']<=lower)
print("Lower Bound:",lower)
print('#Outlier below lower bound:', lower_array.sum())

# Create arrays of Boolean values indicating the outlier rows
upper_array = np.where(product_catalogue['Price']>=upper)[0]
lower_array = np.where(product_catalogue['Price']<=lower)[0]
print(lower_array)
```

```python
product_catalogue
```

```python
# Removing the outliers
#product_catalogue.drop(index=upper_array, inplace=True)
product_catalogue.drop(index=lower_array, inplace=True)

# Print the new shape of the DataFrame
print("New Shape: ", product_catalogue.shape)
```

```python
# save the cleaned product catalogues
cleaned_product_catalogue = product_catalogue
product_catalogue.to_csv('ratususu_products_cleaned.csv')
```

# Importance of Data Quality

Study case:
The simple solution outputs product name that will achieve the optimum revenue for shop “Ratu Susu”.

We will use different input here (raw and cleaned data) to simulate the importance of data quality for model reliability and performance.

```python
# Raw data as input to the model

product_catalogue['Revenue'] = product_catalogue['Units Sold'] *product_catalogue['Price']
total_revenue = product_catalogue.groupby('Product Name')['Revenue'].sum()
optimum_product_raw = total_revenue.idxmax()
print("Optimum product with raw data input:",  optimum_product_raw )
```

```python
# Cleaned data as input to the model

cleaned_product_catalogue['Revenue'] = cleaned_product_catalogue['Units Sold'] *cleaned_product_catalogue['Price']
total_revenue = cleaned_product_catalogue.groupby('Product Name')['Revenue'].sum()
optimum_product_clean = total_revenue.idxmax()
print("Optimum product with cleaned data input:",  optimum_product_clean)
```
